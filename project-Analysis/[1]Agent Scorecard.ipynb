## Methods for Conceptual Analysis
- Frequency (No of claims) -- This measures how often claims are made. A higher number of claims indicates greater frequency, which could affect profitability negatively.

- Severity (GWP, GIC, NOP_T) -- Severity looks at the financial impact of those claims. For example, you could look at how much the agent brings in (GWP) compared to how much is paid out (GIC). High claims severity with low GWP can indicate lower profitability.

- Calculation:
**Frequency = NOC_TOTAL / NOP_T (Number of claims per policy).**

  
**Severity = GIC / NOC_TOTAL (Average claim cost).**

- **Loss Ratio = GIC/GWP**

  
This is one of the most critical indicators of profitability for any insurance company or agent. A loss ratio above 1 (100%) means the agent is generating more claims payouts than premium income, which is unprofitable. A lower loss ratio (typically below 0.6 or 60%) suggests a profitable agent.

- **Profitability per Policy -- GWP - GIC for each policy**

  
The total GWP collected minus the total claims paid (GIC) for each policy type (ACT vs NON-ACT). Aggregate this across policies to measure total profitability for the agent.

- **Agent's profit margin -- (GWP-GIC)/GWP**

  
This will give you the proportion of the premium income that remains as profit after claims are paid.

- **Average Premium Per Policy -- GWP/NOP_T**

  
to assess how much GWP is generated on average for each policy. A higher average premium per policy might be a sign that the agent is more successful in selling higher-value policies, which could translate to higher profitability.

- **Renewal rate -- no of renewals/NOP_T** (using xc_veh_renewal_no).

Agents with higher renewal rates may demonstrate more consistent profitability because renewing policies generally incur lower acquisition costs.

- **Policy_duration -- expiry_date - inception_date_final**

- **Risk exposure and risk-based profitability** -- using (RISK_NO VS NOP_T) Agents with higher exposure to high-risk policies may have lower profitability, especially if the claims are high.

```python
af['inception_date_final'] = pd.to_datetime(af['inception_date_final'], errors='coerce')
af['inception_year'] = af['inception_date_final'].dt.year
agentYearwise = af.groupby(['agent_name', 'inception_year']).agg({
    'policy_no': 'nunique',
    'split': 'sum',
    'gwp': 'sum',
    'gic': 'sum',
    'noc_total': 'sum',
    'nop_t': 'sum',
    'risk_no': 'sum',
    'branch': 'first',   
    'vehicle_age': 'max', 
    'xc_veh_renewal_no': 'max',
    'policy_count': 'first'
}).reset_index()
print(agentYearwise.shape)
agentYearwise.head()
```

```python
# Vectorized normalize function
def normalize_columns(col):
    min_val, max_val = col.min(), col.max()
    if min_val == max_val:
        return pd.Series([np.nan] * len(col), index=col.index)
    return (col - min_val) / (max_val - min_val)


# Profitability Calculation
def calculate_profitability(data):
    data['Ngwp'] = normalize_columns(data['gwp'])
    data['Nprofit_margin'] = normalize_columns(data['profit_margin'])
    data['Ntotal_profitability'] = normalize_columns(data['total_profitability'])
    return data
# Retention Calculation
def calculate_retention(data):
    data['Nrenewal_rate'] = normalize_column(data['renewal_rate'])
    return data

# Claims Calculation
def calculate_claims(data):
    # Vectorized claims score calculation
    data['Nloss_ratio'] = normalize_column(data['loss_ratio'])
    data['Nfrequency'] = normalize_column(data['frequency'])
    data['Nseverity'] = normalize_column(data['severity'])
    
    # Recalculate claims score with the new weighting (severity/frequency and loss ratio)
    loss_ratio = data['Nloss_ratio']
    frequency = data['Nfrequency']
    severity = data['Nseverity']
    
    claims_score = 1 - (0.4 * loss_ratio + 0.3 * frequency + 0.3 * severity)
    
    # In case the loss ratio is high (e.g., > 0.75), apply an adjusted claims score
    claims_score[loss_ratio > 0.75] = 1 - (0.4 * loss_ratio[loss_ratio > 0.75] + 
                                           0.3 * frequency[loss_ratio > 0.75] + 
                                           0.3 * severity[loss_ratio > 0.75]) * 1.5
    
    data['claims_score'] = claims_score
    return data

# GWP Growth Calculation (vectorized)
def calculate_gwp_growth(data):
    # Sort by agent and policy, then by inception_date_final
    data = data.sort_values(by=['agent_name','inception_year'])
    
    # Calculate previous year's GWP (same agent, same policy)
    data['prev_gwp'] = data.groupby(['agent_name', 'policy_no'])['gwp'].shift(1)
    
    # Calculate growth, taking previous year's GWP (handle NaN by filling it with 0)
    data['gwp_growth'] = (data['gwp'] - data['prev_gwp']) / data['prev_gwp']
    data['gwp_growth'].fillna(0, inplace=True)  # Fill NaN values (first row or no previous year data)
    return data
    
# Final Score Calculation (vectorized)
def calculate_final_score(data):
    # New weighted formula based on the mentor's advice:
    # 40% - Loss ratio & Profit margin
    # 30% - Frequency, Severity, Claims
    # 30% - Growth, Volume, Renewals
    base_score = (
        0.4 * (data['Nprofit_margin'] + data['Nloss_ratio']) +
        0.3 * (data['Nfrequency'] + data['Nseverity'] + data['claims_score']) +
        0.15 * (data['gwp_growth'] + data['Nrenewal_rate'])
    )
    
    # Apply penalty: If loss ratio > 0.75, subtract a fixed penalty from base score
    penalty = (data['loss_ratio'] > 0.75) * 0.25  # you can tune this value
    data['final_score'] = base_score - penalty

    # Ensure scores don't drop below 0
    data['final_score'] = data['final_score'].clip(lower=0)

    # **New Step**: Do not force "Very Low Tier" to 0, but retain the actual final score for these agents
    data['final_score'] = data.apply(
        lambda row: row['final_score'] if not (row['gwp'] <= 10000 and row['gic'] < 7500 and row['loss_ratio'] > 0.75) else row['final_score'],
        axis=1
    )
    return data

def classify_tiers(data):
    # Step 1: Label as "VERY Low Tier" for the forced condition (even if the score is higher)
    #         This step ensures that the tier is determined by the conditions, not the score
    data['tier'] = np.select(
        [
            (data['gwp'] <= 10000) & (data['gic'] < 7500) & (data['loss_ratio'] > 0.75),  # VERY Low Tier condition
            data['final_score'] > 0.75,  # Top Tier
            (data['final_score'] >= 0.5) & (data['final_score'] <= 0.75)  # Mid Tier
        ],
        ['VERY Low Tier', 'Top Tier', 'Mid Tier'],
        default='Low Tier'  # Low Tier as default if none of the above conditions match
    )
    
    # Step 2: Handle NaN values for final_score (if any)
    data.loc[data['final_score'].isna(), 'tier'] = 'NaN'

    # **Step 3**: Optional: Force Low Tier for extreme high loss ratios, if necessary
    # data.loc[data['loss_ratio'] > 0.80, 'tier'] = 'Low Tier'  # Uncomment if needed
    return data
```

```python
def process_agent_scores_optimized(data):
    # Vectorized operations for profitability, retention, claims, etc.
    data = calculate_profitability(data)
    data = calculate_retention(data)
    data = calculate_claims(data)
    data = calculate_gwp_growth(data)
    data = calculate_final_score(data)
    data = classify_tiers(data)
    return data

# Apply the function to the data
data = agentYearwise.copy()
data = process_agent_scores_optimized(data)

# Display the final DataFrame with agent scores and tiers
print(data[['agent_name', 'final_score', 'tier']].head())
```

# ***Final score logic***

```python
# Thresholds for each line of business (lob)
lob_thresholds = {
    'PAC': 0.50,
    'FIR': 0.50,
    'LIA': 0.50,
    'ENG': 0.60,
    'WRK': 0.20,
    'ACC': 0.45,
    'MAR': 0.35,
    'MED': 0.65,
    'BON': 0.10,
    'HUL': 0.65,
    'MOT': 0.75
}
```

```python
# Weights
WEIGHT_PROFITABILITY = 0.40
WEIGHT_CLAIMS = 0.30
WEIGHT_GWP_GROWTH = 0.30

# Normalize column
def normalize_columns(col):
    min_val, max_val = col.min(), col.max()
    return (col - min_val) / (max_val - min_val) if min_val != max_val else pd.Series([np.nan]*len(col), index=col.index)

# Profitability calculation
def calculate_profitability(data):
    def assign_lob_score(row):
        threshold = lob_thresholds.get(row['lob'], 0.75)
        return threshold - row['loss_ratio']
    
    data['lob_score'] = data.apply(assign_lob_score, axis=1)
    data['Nlob_score'] = normalize_columns(data['lob_score'])
    return data

# Claims score (no penalty)
def calculate_claims(data):
    data['severity'] = data['severity'].fillna(0)
    data['frequency'] = data['frequency'].fillna(0)
    data['claims_score'] = data['frequency'] + data['severity']
    data['Nclaims_score'] = normalize_columns(data['frequency'] + data['severity'])
    return data

# GWP growth calculation
def calculate_nep_growth(data):
    data = data.sort_values(by=['agent_name', 'lob', 'account_yr'])
    data['prev_nep'] = data.groupby(['agent_name', 'lob'])['nep'].shift(1).fillna(0)
    data['nep_growth'] = ((data['nep'] - data['prev_nep']) / data['prev_nep'].replace(0, np.nan)).fillna(0)
    data['Nnep_growth'] = normalize_columns(data['nep_growth'])
    return data

# Final score
def calculate_final_score(data):
    data['final_score'] = (
        WEIGHT_PROFITABILITY * data['Nlob_score'] +
        WEIGHT_CLAIMS * data['Nclaims_score'] +
        WEIGHT_GWP_GROWTH * data['Nnep_growth']
    )
    return data

# Main pipeline
def process_agent_score(data, lob_thresholds):
    data = calculate_profitability(data)
    data = calculate_claims(data)
    data = calculate_nep_growth(data)
    data = calculate_final_score(data)
    return data

# Assuming N is your DataFrame
data = N.copy()
# data['Nfrequency'] = normalize_columns(data['frequency'])
# data['Nseverity'] = normalize_columns(data['severity'])
data = process_agent_score(data, lob_thresholds)

# Show results
data[['account_yr', 'agent_name', 'lob','nep','nic','nop_t','noc', 'lob_score','frequency','severity','nep_growth','final_score']].head()
```

## NOTE: CONDITIONS AND ASSUMPTIONS FOR THE FUNCTIONS ABOVE--

1. ### normalize_columns(col)
   - Purpose: nromalize a numeric column between 0 and 1
   - if all values in col are the same, it will return NaN or produce a divide-by-zero (not handled explicitly).
    value -min / max - min

2. ### calculate_profitability(data)
    - Purpose: normalize profitability-related metrics
    - gwp -> Ngwp, profit_margin -> Nprofit_margin, total_profitability 0-> Ntotal_profitability

3. ### calculate_claims(data, lob_thresholds)
    - purpose: compute normalized claim score adjusted for LOB-specifc thresholds
    - NaN values in 'severity' should be treated as 0.
    - there is lob_threshods dict that maps LOB to max acceptable loss ratios
    - penalty is applied if loss> threshold
    - raw claims score = 1 - (0.4* loss + 0.3 * freq + 0.3 * sev)
    - for each row:
    - if loss_ratio > threshold, apply 1.5x penalty to the weighted formula ;  else retain the score.

4. ### calculate_gwp_growth(data)
    - purpose: calculate year-over-year GWP growth by agent and LOB
    - growth is computed related to previous year for the same agent & LOB
    - NaN or missing previous gwp values are treated as 0% growth
    - basically, sort data by agent_name and account year
    - shift gwp to get prev_gwp per agent & lob
    - compute current - previous / previous
    - fill NaNs (first year entries) with 0
    
5. ### calculate_final_score(data, lob_thresholds)
    - purpose: aggregate all metrics into a weighted final score with penalties, with special handling for VERY LOW TIER
    - scoring formula:
      40% = loss ratio + profit margin

      30% = frequency, severity, claims score

      30% = gwp growth (volume-related metric)
    - GIC IS USED TO OVERRIDE FOR VERY LOW TIER
    - if gwp <10,000 and gic exceed the lobThreshold x gwp - > set score to 0 "very low tier"
    - final scores cant be -ve so clip to 0
      
6. ### classify_tiers(data, lob_thresholds)
    - purpose: assign each record a tier based on final score and special conditions
    - Tiering is hierarchical:
    - 1. VERY LOW TIER --- overrides all based on [gwp < 10000 and gic > threshold * gwp]
      2. TOP TIER --- final score >0.75
      3. MID TIER --- final score b/w 0.5 and 0.75
      4. LOW TIER --- everything else
    - if final_score is NaN, assign 'NaN' as tier.
7. ### process_agent_Score(data, lob_thresholds)
    - Normalize profitability metrics.
    - Calculate claims score (with penalty).
    - Compute GWP growth and normalize.
    - Aggregate into a final weighted score.
    - Assign tiers.

```python
def classify_agents(df, score_col='final_score', ratio_col='nic/nep ratio',
                    score_pct=0.75, ratio_pct=0.70):
    """
    Classifies agents into 4 quadrants based on final_score and NIC/NEP ratio.

    Parameters:
    - df: DataFrame containing agent performance data
    - score_col: Column name for final_score
    - ratio_col: Column name for NIC/NEP ratio
    - score_pct: Percentile threshold for high final_score (default = 75th)
    - ratio_pct: Percentile threshold for high NIC/NEP ratio (default = 65th)

    Returns:
    - DataFrame with additional columns: 'score_level', 'ratio_level', 'agent_class'
    """
    # Replace 0s or negatives in ratio with small positive value for safety
    df[ratio_col] = df[ratio_col].clip(lower=0)

    # Calculate thresholds
    score_thresh = df[score_col].quantile(score_pct)
    ratio_thresh = df[ratio_col].quantile(ratio_pct)

    # Assign High/Low levels
    df['score_level'] = np.where(df[score_col] >= score_thresh, 'High', 'Low')
    df['ratio_level'] = np.where(df[ratio_col] >= ratio_thresh, 'High', 'Low')

    # Map quadrant labels
    def assign_class(row):
        if row['score_level'] == 'High' and row['ratio_level'] == 'Low':
            return 'Efficient Performer'
        elif row['score_level'] == 'High' and row['ratio_level'] == 'High':
            return 'High-Loss Performer'
        elif row['score_level'] == 'Low' and row['ratio_level'] == 'High':
            return 'High-Risk Underperformer'
        else:
            return 'Low Performer'

    df['agent_class'] = df.apply(assign_class, axis=1)

    return df

# def label_quadrant(group):
#     fs_thresh = group['final_score'].median()
#     ratio_thresh = group['nic'] / group['nep']
#     group['nic_nep_ratio'] = ratio_thresh

#     ratio_median = group['nic_nep_ratio'].median()

#     def assign_zone(row):
#         score = row['final_score']
#         ratio = row['nic_nep_ratio']
#         if score >= fs_thresh and ratio < ratio_median:
#             return 'Prime Performer'
#         elif score < fs_thresh and ratio < ratio_median:
#             return 'Underutilized Agent'
#         elif score >= fs_thresh and ratio >= ratio_median:
#             return 'Costly Contributor'
#         else:
#             return 'High-Risk Liability'

#     group['quadrant'] = group.apply(assign_zone, axis=1)
#     return group
```
